```python
import pandas as pd

# Load the data to inspect its contents
file_path = 'processed_text_data_001.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Plotting the distribution of race_of_household
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='race_of_household')
plt.title('Distribution of Race of Household')
plt.xlabel('Race')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting the distribution of race across different streets
plt.figure(figsize=(12, 8))
sns.countplot(data=data, x='street', hue='race_of_household', palette='Set2')
plt.title('Race Distribution by Street')
plt.xlabel('Street')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Race of Household', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
